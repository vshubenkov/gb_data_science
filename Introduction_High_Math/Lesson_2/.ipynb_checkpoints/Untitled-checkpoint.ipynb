```python
def function_elipce(a, b):
    x1 = np.array([])
    x = np.linspace(100,-100,10000)
    for i in x:
        if (1 - (i ** 2)/a >= 0):
            x1 = np.append(x1,i)
    y = np.sqrt((1 - (x1 ** 2) / a ) * b)
    plt.axis([-10, 10, -10, 10])
    plt.axis('square')

    title_dict = {'fontsize':10, 'fontweight':'bold','color':'Black'}
    title_legend = {'size':10}

    plt.plot(x1, y, label='x',color='black')
    plt.plot(x1, -y, label='x',color='black',linestyle='dashed')
    
    #legend= plt.legend(loc='lower right', prop=title_legend, frameon=False)
    #plt.setp(legend.get_texts(),color='grey')

    plt.grid(axis='y',color='grey')
    plt.grid(axis='x',color='grey')
```

```python
function_elipce(10,15)
```

```python
def function_hyperbole(a, b):
    x1 = np.array([])
    x = np.linspace(30,0,2000)
    for i in x:
        if (((i ** 2)/a) - 1>= 0):
            x1 = np.append(x1,i)
    y = np.sqrt((((x1 ** 2) / a ) - 1) * b)
    
    plt.axis([-10, 10, -10, 10])
    title_dict = {'fontsize':10, 'fontweight':'bold','color':'Black'}
    title_legend = {'size':10}
    plt.plot(x1, y, label='x',color='black')
    plt.plot(x1, -y, label='x',color='black',linestyle='dashed')
    plt.plot(-x1, y, label='x',color='black')
    plt.plot(-x1, -y, label='x',color='black',linestyle='dashed')
    #legend= plt.legend(loc='lower right', prop=title_legend, frameon=False)
    plt.setp(legend.get_texts(),color='grey')
    plt.grid(axis='y',color='grey')
    plt.grid(axis='x',color='grey')
    return y, x1
```

```python
Y, X = function_hyperbole(6,5)
```

```python
def function_parabolla_3D(a, b):
    x = np.linspace(4,-4,100)
    y = x
    fig = figure()
    ax = Axes3D(fig)  
    x, y = np.meshgrid(x,y)
    z = 1 - x ** 2/a ** 2 -  y ** 2/b ** 2 
    ax.plot_wireframe(x,y,z)
    ax.scatter(0,0,0,'z',50,'red')
    show()
```
