```python
def Polar_cycle(R):
    ungel = np.linspace(0,6.28,10)
    
    x = R*np.cos(ungel);
    y = R*np.sin(ungel);
    
    plt.axis('square')
    plt.axis([-10, 10, -10, 10])
    plt.plot(x, y, label='x',color='black')
```

```python
def Polar_line(ungel):
    x = np.linspace(0,10,10)
    y = np.linspace(0,10,10)
    
    x1 = np.sqrt(x*x + y*y)*np.cos(ungel);
    y1 = np.sqrt(x*x + y*y)*np.sin(ungel);
    
    plt.axis('square')
    plt.axis([-10, 10, -10, 10])
    plt.plot(x1, y1, label='x',color='black')
```

```python
def adjust_coordinate(x, y, angel, abcisa, ordinata): 
    if angel < 0:
        alfa = 180
        x1 = x*np.cos(np.deg2rad(alfa)) + y*np.sin(np.deg2rad(alfa))
        y1 = -x*np.sin(np.deg2rad(alfa)) + y*np.cos(np.deg2rad(alfa))                                                  
    return x1, y1
        
        
```
