#Перевод полярных координат в декартовы


```python
def pol_dec(R,fi):
    x1 = R*np.cos(fi)
    y1 = R*np.sin(fi)
    return x1, y1

def dec_pol(x, y):
    R = np.sqrt(x**2 + y**2)
    tan_fi = y/x
    return R, np.arctan(tan_fi)*180/np.pi
```

```python
print(pol_dec(2.8284271247461903,0.7853981633974483))
```

```python
print(dec_pol(2,2))
```
