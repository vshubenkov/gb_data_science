```python
title_dict = {'fontsize':16, 'fontweight':'bold','color':'DarkRed'}
plt.plot(x,x2)
plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('Xline',fontdict=title_dict)
plt.ylabel('Yline',fontdict=title_dict)


plt.show()
```

```python
x3 = x**3
```

```python
title_dict = {'fontsize':16, 'fontweight':'bold','color':'DarkRed'}
plt.plot(x,x2)
plt.plot(x,x3)
plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('X-line',fontdict=title_dict)
plt.ylabel('Y-line',fontdict=title_dict)


plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

title_dict = {'fontsize':16, 'fontweight':'bold','color':'DarkRed'}

plt.plot(x,x2)
plt.plot(x,x3)

plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('X-line',fontdict=title_dict)
plt.ylabel('Y-line',fontdict=title_dict)

plt.legend(labels=['x2','x3'])

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

title_dict = {'fontsize':16, 'fontweight':'bold','color':'DarkRed'}

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('X-line',fontdict=title_dict)
plt.ylabel('Y-line',fontdict=title_dict)

plt.legend(loc='lower right')

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

title_dict = {'fontsize':16, 'fontweight':'bold','color':'DarkRed'}

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('X-line',fontdict=title_dict)
plt.ylabel('Y-line',fontdict=title_dict)

plt.legend(loc='lower right', fontsize = 8)

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

title_dict = {'fontsize':10, 'fontweight':'bold','color':'DarkRed'}
title_legend = {'size':5}

plt.plot(x, x2, label='x**2')
plt.plot(x, x3, label='x**3')

plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('X-line',fontdict=title_dict)
plt.ylabel('Y-line',fontdict=title_dict)

plt.legend(loc='lower right', prop=title_legend)

plt.show()
```

```python
plt.axis([-5, 5, -8, 8])

title_dict = {'fontsize':10, 'fontweight':'bold','color':'DarkRed'}
title_legend = {'size':10}

plt.plot(x, x2, label='x**2',color='red',linestyle='dashed')
plt.plot(x, x3, label='x**3',color='green',linestyle='dotted')

plt.title('Grafic f(x)', fontdict=title_dict, loc='left')
plt.xlabel('X-line',fontdict=title_dict)
plt.ylabel('Y-line',fontdict=title_dict)

legend= plt.legend(loc='lower right', prop=title_legend, frameon=False)
plt.setp(legend.get_texts(),color='grey')

plt.grid(axis='y',color='grey')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-5,5,101)
x2 = x**2
x3 = x**3
```

```python
plt.axis([-5,5,-8,8])#border
plt.plot(x,x2)
plt.plot(x,x3)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1)
ax.plot(x, x2)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1)
print(ax.shape)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
plt.subplots_adjust(hspace=0.25)
```

```python
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(nrows=2,ncols=2)
print(ax.shape)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
